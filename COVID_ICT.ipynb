```python
# import libraries 
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import random
import seaborn as sns
from fbprophet import Prophet
```

```python
covid=pd.read_csv("Daily_Cases_By_Province.csv")
covid.columns.values
```

```python
covid_ICT=covid[['Date','ICT']]
covid_ICT=covid_ICT.rename(columns={'Date':'ds','ICT':'y'})
covid_ICT['y']=covid_ICT['y'].diff()
covid_ICT['y']=covid_ICT['y'].fillna(0)
```

```python
model=Prophet()
model.fit(covid_ICT)
month = model.make_future_dataframe(periods=30)
forecast_month = model.predict(month)
```

```python
month_figure=model.plot(forecast_month,xlabel='Date',ylabel='Cases Per Day')
```
